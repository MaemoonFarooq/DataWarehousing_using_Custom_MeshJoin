```python
# Hash map size
hash_map_size = 100

# First hash function (different formula)
def hash_function_1(value):
    return (value * 3 + 7) % hash_map_size

# Second hash function (different formula)
def hash_function_2(value):
    return ((value + 5) * 2) % hash_map_size

# Create an array of values from 0 to 101
values = list(range(102))

# Track occupied hash map indices
occupied = set()

# Test both hash functions on the same array
print("Value | Hash 1 | Hash 2")
print("-" * 25)
for value in values:
    h1 = hash_function_1(value)
    h2 = hash_function_2(value)

    # Fill with Hash 1 if not already occupied
    if h1 not in occupied:
        print(f"{value:5} | {h1:6} | {'-':6}")
        occupied.add(h1)
    # Fill with Hash 2 if Hash 1 already occupied the index
    elif h2 not in occupied:
        print(f"{value:5} | {'-':6} | {h2:6}")
        occupied.add(h2)
    else:
        # If both indices are occupied, print '-'
        print(f"{value:5} | {'-':6} | {'-':6}")
```

```python
# # Sample productid and customerid arrays (you can replace these with actual data)
# productid = [66, 46, 12, 47, 38, 81, 2, 17, 95, 86, 75, 80, 45, 4, 34, 85, 66, 52, 15, 19, 4, 76, 53, 6, 100, 27, 30, 22, 88, 59]
# customerid = [17, 50, 94, 80, 20, 85, 55, 45, 67, 23, 27, 82, 4, 12, 2, 48, 69, 67, 37, 76, 64, 75, 53, 66, 71, 34, 54, 82, 59, 100]

# Define hash map size (at least twice the size of the arrays)
hash_map_size = 60  # You can adjust this based on the total number of values you have

# First hash function (maps to first half of the hash map)
def hash_function_1(value):
    return (value * 3 + 7) % (hash_map_size // 2)  # First half: 0 to 29

# Second hash function (maps to second half of the hash map)
def hash_function_2(value):
    return ((value + 5) * 2) % (hash_map_size // 2) + (hash_map_size // 2)  # Second half: 30 to 59

# Create a hash map (dictionary) to store the values with chaining (lists to handle collisions)
hash_map = {}

# Function to insert value with chaining (handling collisions)
def insert_to_hash_map(value, hash_function, array_name):
    index = hash_function(value)
    if index not in hash_map:
        hash_map[index] = []
    hash_map[index].append((value, array_name))

# Insert values from productid using hash_function_1 and customerid using hash_function_2
for value in productid:
    insert_to_hash_map(value, hash_function_1, 'productid')

for value in customerid:
    insert_to_hash_map(value, hash_function_2, 'customerid')

# Output the results with index-wise comparison
print("Comparison of Hash Functions (Index-wise):")
print(f"{'Index':<6}{'Product ID Value':<15}{'Hash Function 1 Index':<25}{'Customer ID Value':<15}{'Hash Function 2 Index'}")
    
# We will display index-wise comparison for both arrays
for i in range(len(productid)):
    value_1 = productid[i]
    value_2 = customerid[i]
    
    index_1 = hash_function_1(value_1)  # Hash for productid
    index_2 = hash_function_2(value_2)  # Hash for customerid
    
    # Get the values at index_1 for productid and index_2 for customerid
    productid_values_at_index = [item[0] for item in hash_map.get(index_1, [])]
    customerid_values_at_index = [item[0] for item in hash_map.get(index_2, [])]
    
    # Convert the lists to string for proper formatting
    productid_values_at_index_str = ', '.join(map(str, productid_values_at_index)) if productid_values_at_index else "None"
    
    # Print the index-wise comparison with formatted output
    print(f"{i:<6}{value_1:<15}{index_1:<25}{productid_values_at_index_str:<15}")
```
